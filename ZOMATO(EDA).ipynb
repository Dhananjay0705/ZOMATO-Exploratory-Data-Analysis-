```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df,df_country,on = 'Country Code',how = 'left')
```

```python
final_df.head(1)
```

```python
## TO CHECK DATA TYPE
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index ## INDEX is use for country name
```

```python
country_val = final_df.Country.value_counts().values
```

```python
## pie chart - Top 3 countries that uses ZOMATO
plt.pie(country_val[:3],labels =country_names[:3],autopct = '%1.2f%%' )
```

```python
# Observation : Zomato maximum records or transaction are from India after that USA and yhen United Kingdom
```

```python
final_df.columns
```

```python
Ratings = final_df.groupby([ 'Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
Ratings
```

 Observation
1. When Rating is between 4.5 - 4.9 -----> Excellent
2. When Rating is between 4.0 - 4.4 -----> Very Good
3. When Rating is between 3.5 - 3.9 ----->Good
4. When Rating is between 2.5 - 3.4 ----->Average
5. When Rating is between 1.8 - 2.4 ----->Poor



```python
Ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x ="Aggregate rating", y ="Rating Count", data = Ratings)
```

```python
sns.barplot(x ="Aggregate rating", y ="Rating Count",hue = 'Rating color', data = Ratings,palette=['white','red','orange','yellow','pink','blue'])
```

**OBSERVATION**

1. Not Rated count is very high
2. Maximum number of rating are   between 2.5 to 3.4

```python
 ## count plot
sns.countplot(x ="Rating color",data = Ratings,palette=['white','red','orange','yellow','pink','blue'])
```
